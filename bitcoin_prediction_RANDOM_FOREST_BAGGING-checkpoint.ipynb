When a person spends a bitcoin, that transaction will need to be verified by a set of computers owned by “miners” in the Bitcoin network. Transaction fees are required in order to have these transactions processed by miners. The spender may specify how much fee to add but miners can only confirm 1MB worth of transactions for each batch or block. If the number of transactions waiting to confirm exceeds what can fit in 1 block, a miner can choose to confirm the transactions with the highest bitcoin fees.

# RANDOM FOREST

To perform a Random Forest training on our dataset, we assign the X and y variables. X will have all the columns that we are assuming to be factors affecting price, y.

```python
features = ['BTC network hashrate','Average BTC block size','NUAU - BTC','Number TX - BTC','TX fees - BTC',"Difficulty - BTC",'Estimated TX Volume USD - BTC','Gold in USD','Litecoin Price','Nasdaq composite index','DJI','Google trends']
X = df[features]
y = df['BTC price']
```

Then we split the dataset into a training and testing or validation dataset. We need this so we can test how good our predictions are to the actual prices.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1)
```

We now feed the training data to the Random Forest algorithm so it can map patterns or decision trees.

```python
model.fit(X_train, y_train)
```

Once we have a model, we can call predict to predict the prices of bitcoin on our testing or validation data. That is, given that we’ve trained a model using X_train and Y_train, we are now interested if it’d be able to ouput a number close to Y_test given X_test. Remember that we split the dataset earlier to a training dataset (X_train, Y_train) and a testing dataset (X_test, y_test).

```python
y_pred = model.predict(X_test)
```

To see the predicted values of the model on the last 5 values of test data, we can call:

```python
print(model.predict(X.head()))
df['BTC price'].head()
```

```python
from sklearn.metrics import mean_absolute_error
rf_val_mae = mean_absolute_error(y_test,y_pred)
rf_val_mae
```

Applying various Evaluation techniques to evaluate the performance of the model

```python
from sklearn import metrics

y_true = y_test # Your real values / test labels
#y_pred = [...] # The predictions from your ML / RF model


print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_true, y_pred))
print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_true, y_pred))
print('Root Mean Squared Error (RMSE):', metrics.mean_squared_error(y_true, y_pred, squared=False))
print('Explained Variance Score:', metrics.explained_variance_score(y_true, y_pred))
print('Max Error:', metrics.max_error(y_true, y_pred))
print('Mean Squared Log Error:', metrics.mean_squared_log_error(y_true, y_pred))
print('Median Absolute Error:', metrics.median_absolute_error(y_true, y_pred))
print('R^2:', metrics.r2_score(y_true, y_pred))
```

Maily the best evaluation techniques for Random Forest Regressor are Root Mean Squared Error(RMSE), Mean Abolute Error(MAE) and R Squared error. They explain the following;
MAE - Mean of Abolute difference between the actual and predicted values
R Squared Error - The percentage of variability explained by the model in the data

MAE score is 0.0075 which is a good score. Same goes for R-squared value

```python
mae = metrics.mean_absolute_error(y_true, y_pred)
rmse = metrics.mean_squared_error(y_true, y_pred, squared=False)
Rsquare = metrics.r2_score(y_true, y_pred)
```
